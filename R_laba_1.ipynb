```r
# Подключаем нужные библиотеки
install.packages("stringr")
install.packages("textstem")
install.packages("ggplot2")
install.packages("rmarkdown")
install.packages("wordcloud2")
install.packages("wordcloud")
install.packages("RColorBrewer")

library(RColorBrewer)
library(wordcloud)
library(wordcloud2)
library(ggplot2)
library(stringr)
library(textstem)
```

```r
text_data <- readLines('E:/work/rut/text_to_voice/voice_to_text.txt')

print(text_data)
```

```r

# работаем с текстом

text <- paste(text_data, collapse = " ") 

text_upper <- toupper(text)

# через регулярки убираем пунктуацию
text_clean <- gsub("[[:punct:]]", "", text_upper)

# разбиваем
words <- unlist(strsplit(text_clean, "\\s+"))  # Разделяем по пробелам

# 5. Лемматизация
words_lem <- lemmatize_words(words)

unique_words <- unique(words_lem)

unique_words
```

```r
# мешок слов в одну строчку!
word_freq <- table(words_lem)

word_freq <- sort(word_freq, decreasing = TRUE)

word_freq
```

```r
df_word_frequencies <- as.data.frame(word_freq)
colnames(df_word_frequencies) <- c("word", "frequency")
```

```r
ggplot(df_word_frequencies[1:10,], aes(x = reorder(word, -frequency), y = frequency)) +
  geom_bar(stat = "identity") +
  labs(title = "Топ-10 слов по частоте", x = "Слова", y = "Частота") +
  theme_minimal() 

```

```r
ggplot(df_word_frequencies, aes(x = frequency)) +
  geom_histogram(binwidth = 1, fill = "blue", color = "black", alpha = 0.7) +
  labs(title = "Распределение частоты слов", x = "Частота", y = "Количество слов") +
  theme_minimal()
```

```r
# для облака нужна отдельная библа
install.packages("wordcloud")
install.packages("RColorBrewer")
library(RColorBrewer)
library(wordcloud)

options(repr.plot.width = 12, repr.plot.height = 8)

set.seed(1234) 
wordcloud(words = df_word_frequencies$word, 
          freq = df_word_frequencies$frequency, 
          min.freq = 3,
          max.words = 50,  
          random.order = FALSE,  # чаще=центр
          rot.per = 0.1,  # поворот
          scale = c(3, 4),  # скейл слов
          colors = brewer.pal(8, "Set3"))
```

```r
# получаем текстовый вектор
df_word_frequencies$word <- as.character(df_word_frequencies$word)

# убираем слова в один символ
df_word_frequencies <- df_word_frequencies[nchar(df_word_frequencies$word) > 1, ]

set.seed(666) 
wordcloud2(data = df_word_frequencies, 
           size = 1.5,  #  масштаб
           shape = 'circle',  # форма
           minRotation = 0,
           maxRotation = 0)

```
